```python
def read_geophone_data_file(filepath, n_columns=49):
    '''Read text file with voltages from geophones.
    
    Input:
    -----
    filepath - Path to file with geophone data
    
    Output:
    ------
    time                     - 1d array with time data (s)
    geophone_data            - 2d (Nx49) array with voltages from geophones. Each column corresponds
                               data from a different geophone for a total of 48 geophones. The last 
                               column has time data.
    normalized_geophone_data - Each column in 'geophone_data' normalized by the maximum reading in that
                               column.
    loc                      - Number to voltage readings. i.e loc = N = Number of rows in output arrays
    
    '''
    
    #Open geophone data file
    geophone_data_file = open(filepath, 'r')

    geophone_data = zeros((20000,n_columns))

    loc=0
    for line in geophone_data_file:
        temp1 = line.split()
    
        for j in range(0,len(temp1)):
            geophone_data[loc,j] = float(temp1[j])
    
        loc += 1
    
    #Extract time from last column in 'vertical_geophone_array' and assign it 
    #to new vector 'time'
    time = zeros(20000)
    time = geophone_data[:,-1]
    
    #Normalize the geophone data
    normalized_geophone_data = zeros((20000,n_columns))

    for j in range(0,n_columns):
        temp2 = max(abs(geophone_data[:,j]))

        for i in range(0,20000):
            normalized_geophone_data[i,j] = geophone_data[i,j]/temp2
    
    geophone_data_file.close()
    
    return time[0:loc], geophone_data[0:loc], normalized_geophone_data[0:loc], loc
```

# Downhole Testing

Details of downhole testing should be carried out are provided in ASTM D7400.

<img src="Figures/Insitu Site Characterization\HW 10\ASTM D7400 - Downhole Travel Paths.png" style="max-width:100%; width: 40%">
<center> <em>Figure 2: Schematic of downhole seismic test in cased boring (ASTM D7400)</em> </center>

The straight-line slant travel distances ($L_R$) can be calculated as follows:
$$L_R = \sqrt{(E_S - E_G + D_G)^2 + X^2}$$

$E_S$ - elevation of the ground surface in contact with the energy source at the center of the energy source, 
$E_G$ - elevation of the top of the geophone borehole,
$D_G$ - depth of the geophone receiver,
$X$   - horizontal distance between the center of the energy source and the geophone borehole/sounding.

```python
def fit_fun_2(t,y, n1=0,n2=1, location=0.0, plot_n1_n2='No', print_eq='No', fit_line_extents=[0,1], ls='b-',
            layer='1', shot_location='?'):

    p = polyfit(t[n1:n2],y[n1:n2],1)
    
    if print_eq=='Yes':
        print 'y(t) = %.5ft %+.5f' %(p[0],p[1])
        
    if plot_n1_n2=='Yes':
        plot([t[n1],t[n2]],[y[n1],y[n2]], 'ks')

    fitted_line = poly1d(p)

    t1 = array(fit_line_extents)
    y1 = fitted_line(t1)
    plot(t1,y1, ls)

    print '%s\t     %s              %8.0f' %(layer, shot_location, p[0])
    
    return p
```

### Travel distance calculations

The borehole deviations for $SD1$ and $SD2$ are given below. The distance between the boreholes need to be corrected by accounting for these deviations at each depth that cross-hole readings were taken. Equation 1 from ASTM D4428 can be used for this purpose:
$$l = \sqrt{[(E_s - D_s)-(E_G - D_G)]^2 + [L\cos \phi + x_G - x_S]^2 + [L \sin \phi + y_G - y_S]^2}$$
Where $\phi$ is the azimuth relative north at the top of the source borehole and the top of the geophone borehole, $L$ is the horizontal distance between the tops of the source and geophone boreholes, $E_S$ and $E_G$ are elevations at the tops of the boreholes, $D_S$ and $D_G$ are depths to the source and the geophone, $x_S$ and $x_G$ are the north deviations of the boreholes at the test depth, $y_S$ and $y_G$ are the east deviations of the boreholes at the test depths. # SCPT - Pseudo Interval Method

<img src="Figures/Insitu Site Characterization\HW 10\ASTM D7400 - SCPT Schematic.png" style="max-width:100%; width: 40%">
<center> <em>Figure 2: Schematic of a single receiver SCPT (ASTM D7400)</em> </center>

The straight-line slant travel distances ($L_R$) can be calculated as follows:
$$L_R = \sqrt{(E_S - E_G + D_G)^2 + X^2}$$

$E_S$ - elevation of the ground surface in contact with the energy source at the center of the energy source, 
$E_G$ - elevation of the top of the geophone borehole,
$D_G$ - depth of the geophone receiver,
$X$   - horizontal distance between the center of the energy source and the geophone borehole/sounding.

Hence, $$L_1 = \sqrt{(E_S - E_G + D_1)^2 + X^2}$$ and $$L_2 = \sqrt{(E_S - E_G + D_2)^2 + X^2}$$

Based on the pseudo-interval metohd, these values can then be used to calculate the average S-wave velocity between $D_1$ and $D_2$. $$V_S = \frac{L_2 - L_1}{T_2 - T_1}$$

```python
cone_tip_depth      = array([5.78,8.99,12.20,15.40,18.61,22.17,25.38,28.57,31.77,34.98,38.43,41.63,44.84,48.04,51.76,
                             57.68,60.88,64.07,67.55,70.75,74.51]) #ft
geophone_depth      = array([3.93,7.14,10.35,13.55,16.76,20.32,23.53,26.72,29.92,33.13,36.58,39.78,42.99,46.19,49.91,
                             55.83,59.03,62.22,65.70,68.90,72.66]) #ft
ray_path            = array([5.97,8.44,11.29,14.28,17.35,20.81,23.96,27.10,30.26,33.43,36.86,40.03,43.22,46.41,50.11,
                             56.01,59.20,62.38,65.85,69.05,72.80,]) #ft
incremental_dist    = array([5.97,2.47,2.85,2.99,3.08,3.46,3.14,3.14,3.16,3.18,3.42,3.18,3.19,3.18,3.70,5.90,3.19,3.18,
                             3.47,3.19,3.75]) #ft
s_wave_arrival_time = array([0.0176,0.0218,0.0271,0.0318,0.0372,0.0455,0.0529,0.0586,0.0622,0.0664,0.0703,0.0752,0.0809,
                             0.0853,0.0899,0.0962,0.0986,0.1018,0.1037,0.1064,0.1087]) #s

N = len(s_wave_arrival_time)

V_s            = zeros(N)
slant_angle    = zeros(N)
corrected_time = zeros(N)
V_s_corrected  = zeros(N)
time_interval  = zeros(N)

for i in range(1,N):
    time_interval[i] = s_wave_arrival_time[i]-s_wave_arrival_time[i-1]
    V_s[i] = incremental_dist[i]/(s_wave_arrival_time[i]-s_wave_arrival_time[i-1])
    corrected_time[i] = s_wave_arrival_time[i]*cos(slant_angle[i])
    V_s_corrected[i]  = incremental_dist[i]/(corrected_time[i]-corrected_time[i-1])
    
    
plot(V_s, geophone_depth, 'o')
plot(V_s_corrected, geophone_depth, 'kx')

ax=gca()
ax.invert_yaxis(), ax.xaxis.tick_top(), ax.xaxis.set_label_position('top')
xlabel('Velocity (m/s)'), ylabel('Depth (m)'), grid(True), legend(loc='lower right', fontsize='small')
```
